# Hill Climbing
### Esse algoritmo começa com uma solução aleatória e tenta melhorar movendo rainhas para reduzir conflitos.

```python
import random

def random_board(n):
    return [random.randint(0, n - 1) for _ in range(n)]

def conflicts(board):
    count = 0
    for i in range(len(board)):
        for j in range(i + 1, len(board)):
            if board[i] == board[j] or \
               abs(board[i] - board[j]) == abs(i - j):
                count += 1
    return count

def hill_climbing(n):
    board = random_board(n)
    while True:
        current_conflicts = conflicts(board)
        if current_conflicts == 0:
            return board
        next_board = list(board)
        for col in range(n):
            best_row = board[col]
            min_conflicts = current_conflicts
            for row in range(n):
                next_board[col] = row
                c = conflicts(next_board)
                if c < min_conflicts:
                    min_conflicts = c
                    best_row = row
            board[col] = best_row

print(hill_climbing(8))
```
