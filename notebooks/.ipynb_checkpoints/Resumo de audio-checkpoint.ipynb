# ===============================
# CONFIGURAÇÕES GERAIS
# ===============================

# Tamanho máximo dos pedaços em milissegundos (5 minutos)
CHUNK_LENGTH_MS = 5 * 60 * 1000

# Modelo de transcrição Whisper
WHISPER_MODEL = "base"  # pode trocar por "small", "medium" ou "large"

# Modelo de sumarização
T5_MODEL = "unicamp-dl/ptt5-base-portuguese-vocab"

device = "cuda" if torch.cuda.is_available() else "cpu"

# ===============================
# FUNÇÃO: Converter para WAV
# ===============================
def convert_to_wav(input_path):
    print(f"\nConvertendo '{input_path}' para WAV...")
    audio = AudioSegment.from_file(input_path)
    wav_path = os.path.splitext(input_path)[0] + "_temp.wav"
    audio.export(wav_path, format="wav")
    print(f"Conversão concluída: '{wav_path}'")
    return wav_path

# ===============================
# FUNÇÃO: Dividir o áudio em pedaços
# ===============================
def split_audio(file_path, chunk_length=CHUNK_LENGTH_MS):
    print(f"\nDividindo áudio '{file_path}' em partes menores...")
    audio = AudioSegment.from_wav(file_path)
    chunks = []
    for i in range(0, len(audio), chunk_length):
        chunk = audio[i:i + chunk_length]
        chunk_path = f"{os.path.splitext(file_path)[0]}_chunk_{i//chunk_length}.wav"
        chunk.export(chunk_path, format="wav")
        chunks.append(chunk_path)
    print(f"Total de {len(chunks)} partes geradas.")
    return chunks

# ===============================
# FUNÇÃO: Transcrever um áudio
# ===============================
def transcribe_audio(file_path, model):
    print(f"\nTranscrevendo '{file_path}'...")
    result = model.transcribe(file_path, fp16=False)
    return result["text"]

```python
# ===============================
# FUNÇÃO: Gerar resumo com T5
# ===============================
def summarize_text(text):
    print("\nIniciando sumarização...")  # Indica o início do processo de resumo

    # Carrega o tokenizador do modelo T5 pré-treinado
    tokenizer = T5Tokenizer.from_pretrained(T5_MODEL)

    # Carrega o modelo T5 para geração de texto e envia para CPU ou GPU
    summarizer = T5ForConditionalGeneration.from_pretrained(T5_MODEL).to(device)

    # Adiciona o prefixo "resuma:" para indicar ao modelo a tarefa de sumarização
    input_text = "resuma: " + text

    # Tokeniza o texto de entrada, limita a 512 tokens e converte para tensor
    inputs = tokenizer.encode(input_text, return_tensors="pt", truncation=True, max_length=512).to(device)

    # Gera o resumo com parâmetros de controle de tamanho e qualidade
    summary_ids = summarizer.generate(inputs, min_length=50, length_penalty=2.0, num_beams=4)

    # Decodifica os tokens gerados em texto legível, ignorando tokens especiais
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary  # Retorna o resumo gerado
```

```python
# ===============================
# FUNÇÃO PRINCIPAL
# ===============================
def process_audio_files(audio_files):
    # Carrega o modelo Whisper para transcrição e envia para CPU ou GPU
    whisper_model = whisper.load_model(WHISPER_MODEL).to(device)

    # Itera sobre cada arquivo de áudio fornecido
    for audio_file in audio_files:
        # Converte o arquivo original para formato WAV
        wav_path = convert_to_wav(audio_file)

        # Divide o áudio WAV em pedaços menores
        chunks = split_audio(wav_path)

        full_transcription = ""  # Inicializa a variável que armazenará toda a transcrição

        # Transcreve cada pedaço de áudio e acumula o texto
        for chunk in chunks:
            text = transcribe_audio(chunk, whisper_model)
            full_transcription += text + " "

        # Exibe a transcrição completa no console
        print("\n--- TRANSCRIÇÃO COMPLETA ---\n")
        print(full_transcription)

        # Gera o resumo da transcrição usando o modelo T5
        summary = summarize_text(full_transcription)

        # Exibe o resumo final no console
        print("\n--- RESUMO ---\n")
        print(summary)
```

# EXECUÇÃO
# ===============================
if __name__ == "__main__":

```python
#!/usr/bin/env python
# coding: utf-8

# ===============================
# IMPORTAÇÃO DE BIBLIOTECAS
# ===============================
import os  # Interação com o sistema de arquivos
import torch  # Biblioteca de machine learning (PyTorch)
import whisper  # Modelo da OpenAI para transcrição de áudio
from pydub import AudioSegment  # Manipulação de arquivos de áudio
from transformers import T5Tokenizer, T5ForConditionalGeneration  # Tokenizador e modelo T5 para sumarização
import re  # Expressões regulares para verificar pontuação final no resumo

# ===============================
# CONFIGURAÇÕES GERAIS
# ===============================
CHUNK_LENGTH_MS = 5 * 60 * 1000  # Define o tamanho máximo de cada pedaço de áudio (5 minutos)
WHISPER_MODEL = "base"  # Modelo Whisper usado para transcrição
T5_MODEL = "unicamp-dl/ptt5-base-portuguese-vocab"  # Modelo T5 para sumarização em português
device = "cuda" if torch.cuda.is_available() else "cpu"  # Usa GPU se disponível, senão CPU

# ===============================
# FUNÇÃO: Converter para WAV
# ===============================
def convert_to_wav(input_path):
    print(f"\nConvertendo '{input_path}' para WAV...")
    audio = AudioSegment.from_file(input_path)  # Carrega o áudio de entrada
    wav_path = os.path.splitext(input_path)[0] + "_temp.wav"  # Define o nome do novo arquivo WAV
    audio.export(wav_path, format="wav")  # Exporta o áudio no formato WAV
    print(f"Conversão concluída: '{wav_path}'")
    return wav_path  # Retorna o caminho do novo arquivo

# ===============================
# FUNÇÃO: Dividir o áudio em pedaços
# ===============================
def split_audio(file_path, chunk_length=CHUNK_LENGTH_MS):
    print(f"\nDividindo áudio '{file_path}' em partes menores...")
    audio = AudioSegment.from_wav(file_path)  # Carrega o áudio WAV
    chunks = []  # Lista para armazenar os pedaços
    base_name = os.path.splitext(file_path)[0]  # Nome base para os arquivos gerados
    for i in range(0, len(audio), chunk_length):
        chunk = audio[i:i + chunk_length]  # Recorta o pedaço
        chunk_path = f"{base_name}_chunk_{i//chunk_length}.wav"  # Define nome do pedaço
        chunk.export(chunk_path, format="wav")  # Salva o pedaço como WAV
        chunks.append(chunk_path)  # Adiciona à lista
    print(f"Total de {len(chunks)} partes geradas.")
    return chunks  # Retorna a lista de pedaços

# ===============================
# FUNÇÃO: Transcrever um áudio
# ===============================
def transcribe_audio(file_path, model):
    print(f"\nTranscrevendo '{file_path}'...")
    result = model.transcribe(file_path, fp16=False)  # Transcreve o áudio usando Whisper
    return result["text"]  # Retorna apenas o texto transcrito

# ===============================
# FUNÇÃO: Gerar resumo com T5 (com verificação de frase completa)
# ===============================
def summarize_text(text):
    print("\nIniciando sumarização com parâmetros dinâmicos e loop de re-geração...")
    tokenizer = T5Tokenizer.from_pretrained(T5_MODEL)  # Carrega o tokenizador
    summarizer = T5ForConditionalGeneration.from_pretrained(T5_MODEL).to(device)  # Carrega o modelo

    input_text_prefix = "resuma: "
    input_tokens = tokenizer.encode(input_text_prefix + text, return_tensors="pt", truncation=True, max_length=512)
    input_token_count = input_tokens.shape[1]  # Conta os tokens de entrada

    # Define limites dinâmicos de tamanho do resumo
    SUMMARY_RATIO_MIN = 0.15
    SUMMARY_RATIO_MAX = 0.25
    ABSOLUTE_MIN_SUMMARY_LENGTH = 50
    ABSOLUTE_MAX_SUMMARY_LENGTH = 300

    # Calcula limites iniciais de geração
    dynamic_max_length_target = int(input_token_count * SUMMARY_RATIO_MAX)
    initial_effective_max_length = max(ABSOLUTE_MIN_SUMMARY_LENGTH, dynamic_max_length_target)
    initial_effective_max_length = min(ABSOLUTE_MAX_SUMMARY_LENGTH, initial_effective_max_length)

    dynamic_min_length_target = int(input_token_count * SUMMARY_RATIO_MIN)
    initial_effective_min_length = max(ABSOLUTE_MIN_SUMMARY_LENGTH - 40, dynamic_min_length_target)
    initial_effective_min_length = max(10, initial_effective_min_length)

    # Ajuste para textos curtos
    if input_token_count < ABSOLUTE_MIN_SUMMARY_LENGTH * 1.5:
        initial_effective_max_length = min(input_token_count + 10, ABSOLUTE_MIN_SUMMARY_LENGTH + 50)
        initial_effective_min_length = min(input_token_count - 10, initial_effective_max_length - 10)
        initial_effective_min_length = max(10, initial_effective_min_length)

    # Garante que max_length > min_length
    if initial_effective_max_length < initial_effective_min_length:
        initial_effective_max_length = initial_effective_min_length + 10

    print(f"Gerando resumo para input de {input_token_count} tokens.")
    print(f"Parâmetros de geração iniciais: max_length={initial_effective_max_length}, min_length={initial_effective_min_length} tokens.")

    # Parâmetros do loop de re-geração
    INCREMENT_STEP = 1
    MAX_RETRIES = 50
    REGENERATION_HARD_MAX_LENGTH = ABSOLUTE_MAX_SUMMARY_LENGTH + 100

    current_max_length = initial_effective_max_length
    current_min_length = initial_effective_min_length
    summary = ""
    retry_count = 0

    # Loop para tentar gerar um resumo que termine corretamente
    while retry_count <= MAX_RETRIES:
        summary_ids = summarizer.generate(
            input_tokens.to(device),
            max_length=current_max_length,
            min_length=current_min_length,
            length_penalty=2.0,
            num_beams=4,
            early_stopping=True
        )
        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

        # Verifica se o resumo termina com pontuação final
        if re.search(r'[.?!]["\']?\s*$', summary.strip()):
            print("  Resumo gerado termina em frase completa. Sucesso!")
            break

        retry_count += 1

        # Se exceder tentativas ou limite de tokens, encerra
        if retry_count > MAX_RETRIES or current_max_length >= REGENERATION_HARD_MAX_LENGTH:
            print(f"  Atingido o limite de tentativas ou comprimento máximo.")
            if not re.search(r'[.?!]["\']?\s*$', summary.strip()):
                 summary = summary.strip() + "..."
            break

        # Aumenta os limites para nova tentativa
        current_max_length += INCREMENT_STEP
        current_min_length = min(current_max_length - 10, current_min_length + int(INCREMENT_STEP / 2))
        current_min_length = max(10, current_min_length)

    return summary  # Retorna o resumo final

# ===============================
# FUNÇÃO PRINCIPAL
# ===============================
def process_audio_files(audio_files):
    whisper_model = whisper.load_model(WHISPER_MODEL).to(device)  # Carrega o modelo Whisper

    for audio_file in audio_files:
        temp_files_to_clean = []  # Lista de arquivos temporários para limpeza

        try:
            wav_path = convert_to_wav(audio_file)
            temp_files_to_clean.append(wav_path)

            chunks = split_audio(wav_path)
            temp_files_to_clean.extend(chunks)

            full_transcription = ""
            for chunk in chunks:
                text = transcribe_audio(chunk, whisper_model)
                full_transcription += text + " "

            print("\n--- TRANSCRIÇÃO COMPLETA ---\n")
            print(full_transcription)

            summary = summarize_text(full_transcription)
            print("\n--- RESUMO FINAL ---\n")
            print(summary)

        except Exception as e:
            print(f"Ocorreu um erro ao processar '{audio_file}': {e}")
        finally:
            # Remove arquivos temporários
            print(f"\nLimpando arquivos temporários para '{audio_file}'...")
            for f_path in temp_files_to_clean:
                if os.path.exists(f_path):
                    os.remove(f_path)
                    print(f"  - Removido: '{f_path}'")
                else:
                    print(f"  - Arquivo não encontrado: '{f_path}'")

# ===============================
# EXECUÇÃO DO SCRIPT
# ===============================
if __name__ == "__main__":
    # Lista de arquivos de áudio para processar
    arquivos = ['audio1.mpeg', 'audio2.mpeg','audio3.mpeg']
    process_audio_files(arquivos)  # Inicia o processamento
```
