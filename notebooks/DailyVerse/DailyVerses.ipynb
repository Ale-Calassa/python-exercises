```python
import pandas as pd
import random
```

```python
def sortear_verso(caminho_csv):
    # Para ler apenas algumas linhas 'Random(Aleatório)' no arquivo
    total_linhas = sum(1 for _ in open(caminho_csv, encoding='utf-8')) - 1 # -1 servirá para ignorar o cabeçalho
    linha_aleatoria = random.randint(1, total_linhas)

    # Para ler apenas uma linha sorteada
    verso = pd.read_csv(caminho_csv, skiprows=lambda x: x != 0 and x != linha_aleatoria)

    # Mostra o versículo(verso)
    linha = verso.iloc[0]
    print(f"{linha['book_name']} {linha['chapter_number']}:{linha['verse_number']} - {linha['verse_text']}")

sortear_verso('kjv.csv')
```
