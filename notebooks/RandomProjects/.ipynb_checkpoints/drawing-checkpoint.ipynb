# Desenhando com Turtle

### Usando a biblioteca turtle para criar padrÃµes visuais simples

```python
import turtle

def drawing():
    tela = turtle.Screen()
    tela.bgcolor("black")
    artista = turtle.Turtle()
    artista.speed(0)
    cores = ["red", "orange", "yellow", "green", "blue", "purple"]

    for i in range(360):
        artista.color(cores[i % 6])
        artista.forward(i)
        artista.left(59)

    tela.mainloop()
drawing()
```
