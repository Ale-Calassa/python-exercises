# Gerador de Frases Inspiradoras

### Um script para exibir frases motivacionais aleatórias toda vez que você o executa.

```python
import random

def motivation():
    frases = [
        "Você é mais forte do que imagina.",
        "Cada passo conta, mesmo os pequenos.",
        "Respire fundo e devagar. Vai passar.",
        "Você merece coisas boas.",
        "A calma é uma superpotência.",
        "A vida é bela, vale a pena viver."
    ]
    print("\nSua frase do momento: ")
    print(random.choice(frases))
motivation()
```
