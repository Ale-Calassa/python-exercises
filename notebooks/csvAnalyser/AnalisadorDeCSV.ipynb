```python
import pandas as pd

def analisar_csv(caminho_arquivo):
    try:
        # LÃª o arquivo CSV
        df = pd.read_csv(caminho_arquivo)
        print("\nArquivo carregado com sucesso!\n")
        
        # Exibe as primeiras linhas
        print("Primeiras linhas do arquivo:")
        print(df.head())

        # Seleciona apenas colunas numÃ©ricas
        colunas_numericas = df.select_dtypes(include='number')

        if colunas_numericas.empty:
            print("\nNenhuma coluna numÃ©rica encontrada.")
            return

        print("\nEstatÃ­sticas bÃ¡sicas:")
        for coluna in colunas_numericas.columns:
            print(f"\nğŸ”¹ Coluna: {coluna}")
            print(f"   Soma: {colunas_numericas[coluna].sum():.2f}")
            print(f"   MÃ©dia: {colunas_numericas[coluna].mean():.2f}")
            print(f"   Contagem: {colunas_numericas[coluna].count()}")

    except FileNotFoundError:
        print("Arquivo nÃ£o encontrado. Verifique o caminho.")
    except pd.errors.EmptyDataError:
        print("Arquivo vazio.")
    except Exception as e:
        print(f"Erro inesperado: {e}")

# Exemplo de uso
if __name__ == "__main__":
    caminho = input("Digite o caminho do arquivo CSV: ")
    analisar_csv(caminho)
```
