"""
Exercício 11 da seção de estrutura sequencial da Python Brasil:
https://wiki.python.org.br/EstruturaSequencial

Faça um Programa que peça 2 números inteiros e um número real. Calcule e mostre:
 1) o produto do dobro do primeiro com metade do segundo;
 2) a soma do triplo do primeiro com o terceiro;
 3) o terceiro elevado ao cubo.

 Apresente os resultados com duas casas decimais

    >>> from secao_01_estrutura_sequencial import ex_11_contas_matematicas
    >>> numeros = ['3.14', '43', '42']
    >>> ex_11_contas_matematicas.input = lambda k: numeros.pop()
    >>> ex_11_contas_matematicas.calcular_formulas()
    O produto do dobro do primeiro com metade do segundo é 1806.00
    A soma do triplo do primeiro com o terceiro é 132.14
    O terceiro elevado ao cubo é 30.96
"""

```python
# Sem a Função .pop()
def calcular_formulas():
    numbers = [
        int(input("Digite o primeiro número Interiro: ")),
        int(input("Digite o segundo número Inteiro: ")),
        float(input("Digite um número Real"))   
    ]
    produto_dobro_com_metade = (2 * numbers[0]) * (numbers[1] / 2)
    soma_triplo_com_terceiro = (3 * numbers[0]) + numbers[2]
    terceiro_ao_cubo = numbers[2] ** 3
    
    print(f"O produto do dobro do Primeiro Nº ({numbers[0]}) com metade do Segundo ({numbers[1]}) é: {produto_dobro_com_metade}")
    print(f"A soma do triplo do primeiro Nº ({numbers[0]}) com o terceiro Nº ({numbers[2]}) é: {soma_triplo_com_terceiro}")
    print(f"O terceiro Nº ({numbers[2]}) elevado ao cudo é: {terceiro_ao_cubo:.2f}")
calcular_formulas()
```

```python
# Com a função .pop()
def calcular_formulas():
    numbers = [
        int(input("Digite o primeiro número inteiro: ")),
        int(input("Digite o segundo número inteiro: ")),
        float(input("Digite um número real: "))
    ]


    num_real = numbers.pop()
    segundo_num = numbers.pop()
    primeiro_num = numbers.pop()
   
    
    produto_dobro_metade = (2 * primeiro_num) * (segundo_num / 2)
    soma_triplo_com_terceiro = (3 * primeiro_num) + num_real
    terceiro_ao_cubo = num_real ** 3

    print(f"O Produto do dobro do Primeiro Nº ({primeiro_num}) com metade do Segundo ({segundo_num}): {produto_dobro_metade}")
    print(f"Soma do triplo de ({primeiro_num}) com ({num_real}): {soma_triplo_com_terceiro}")
    print(f"{num_real} elevado ao cubo: {terceiro_ao_cubo:.2f}")

calcular_formulas()
```
