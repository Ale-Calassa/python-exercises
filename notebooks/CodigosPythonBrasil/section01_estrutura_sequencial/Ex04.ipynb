# Python Brasil Exercício 04 Notas Bimestrais

"""
Exercíci 04  da seção de estrutura sequencial da Python Brasil:
https://wiki.python.org.br/EstruturaSequencial

Faça um Programa que peça as 4 notas bimestrais e mostre a média.

    >>> from secao_01_estrutura_sequencial import ex_04_notas_bimestrais
    >>> numeros =['7', '8','9','10']
    >>> ex_04_notas_bimestrais.input = lambda k: numeros.pop()
    >>> ex_04_notas_bimestrais.calcular_media()
    A média aual é 8.5

"""

```python
def calcular_media():
    nota01 = float(input("Digite a primeira nota: "))
    nota02 = float(input("Digite a segunda nota: "))
    nota03 = float(input("Digite a terceira nota: "))
    nota04 = float(input("Digite a quarta nota: "))
    media = (nota01 + nota02 + nota03 + nota04) / 4
    print(f"A média anual é: {media:.02f}")
calcular_media()
```
